```python
#Creating a function to fix CÉDULA
def fix_cedula(y):
  if re.match("\d{2}-\d{3}", y) is None:
    hyphen_position = y.find('-')-1
    cedula_fixed = y[:hyphen_position] + "-" + y[hyphen_position] + y[hyphen_position+2:]
    return cedula_fixed 
  else:
    return y
```

```python
#Fixing a the column CÉDULA
homicides_df['CÉDULA'] = homicides_df['CÉDULA'].apply(fix_cedula)
```

```python
# Creating a function to fix CORREO
def fix_correo(y):
  if re.match("^[a-zA-Z0-9]+\d{4}@[a-zA-Z0-9]+\.(edu\.co|com)$", y) is None and re.match("^[a-zA-Z0-9]+\d{4}@[a-zA-Z0-9]+\.gov\.co$", y) != None:
    gov_position = y.find('.gov.')
    correo_fixed = y[:gov_position] + ".edu." + y[gov_position+5:]
    return correo_fixed 
  else:
    return y
```

```python
# Fixing a the column CORREO
homicides_df['CORREO'] = homicides_df['CORREO'].apply(fix_correo)
```

```python
# Creating a function to fix DIA
def fix_day (x):
  week_days = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
  distances = []

  result_t = [k for k in range(1,6)]
  distances = [pylev.levenshtein(x, week_days[i]) for i in range(0,len(week_days))]
  position = pd.Series(distances).idxmin()
  fixed_day = week_days[position]
  
  return week_days[position]
```

```python
# Fixing a the column DIA
homicides_df1['DIA'] = homicides_df1['DIA'].apply(fix_day)
```

```python
from datetime import date

def fix_ano_nacimiento (x):
  current_year = date.today().year
  fixed_year = current_year - x

  return int(fixed_year)
```

```python
# Fixing a the column AÑO DE NACIMIENTO
homicides_df1['AÑO DE NACIMIENTO'] = homicides_df1['EDAD'].apply(fix_ano_nacimiento)
homicides_df1.head()
```
